```python
# SLIDER MATPLOTLIB

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
from math import sin, cos, radians

def vent(V, intx, inty, intt, n, dt=1.0):
    """
    Champ de vent avec slider temps et pas dt, avec pré-calcul pour un affichage rapide.
    
    Arguments :
    - V : fonction qui prend ([x, y], t) et retourne (angle_deg, vitesse)
    - intx : intervalle en x [xmin, xmax]
    - inty : intervalle en y [ymin, ymax]
    - intt : intervalle en temps [tmin, tmax]
    - n : nombre de divisions par axe
    - dt : pas de temps pour le slider
    """

    # Grille
    x_vals = np.linspace(intx[0], intx[1], n + 1)
    y_vals = np.linspace(inty[0], inty[1], n + 1)
    X, Y = np.meshgrid(x_vals, y_vals)
    points = np.column_stack([X.ravel(), Y.ravel()])

    # Liste des temps pour pré-calcul
    ts = np.arange(intt[0], intt[1] + dt, dt)

    # Pré-calculer tous les vecteurs et couleurs
    U_cache = []
    V_cache = []
    C_cache = []

    for t in ts:
        results = np.array([V(p, t) for p in points])  # shape (N_points, 2)
        angles_deg = results[:, 0]
        vitesses = results[:, 1]
        angles_rad = np.radians(angles_deg)
        U = (-vitesses * np.sin(angles_rad)).reshape(X.shape)
        Vv = (-vitesses * np.cos(angles_rad)).reshape(Y.shape)
        C = 1 - np.clip(vitesses / 30, 0, 1)
        U_cache.append(U)
        V_cache.append(Vv)
        C_cache.append(C.flatten())

    # Figure
    fig, ax = plt.subplots()
    plt.subplots_adjust(bottom=0.25)

    # Affichage initial
    Q = ax.quiver(X, Y, U_cache[0], V_cache[0], color=plt.cm.gray(C_cache[0]))
    ax.set_aspect('equal')
    ax.set_title(f"Temps = {ts[0]:.2f}")

    # Slider
    ax_slider = plt.axes([0.25, 0.1, 0.5, 0.03])
    slider = Slider(ax_slider, 'Temps', intt[0], intt[1], valinit=ts[0], valstep=dt)

    # Fonction de mise à jour
    def update(val):
        # Trouver l'index correspondant au temps
        idx = int(round((slider.val - intt[0]) / dt))
        idx = np.clip(idx, 0, len(ts) - 1)
        Q.set_UVC(U_cache[idx], V_cache[idx])
        Q.set_color(plt.cm.gray(C_cache[idx]))
        ax.set_title(f"Temps = {ts[idx]:.2f}")
        fig.canvas.draw_idle()

    slider.on_changed(update)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

def vent(V, intx, inty, intt, n, dt, routes, route_times):
    """
    Champ de vent avec slider temps + affichage routes.

    Arguments :
    - V : fonction ([x, y], t) -> (angle_deg, vitesse)
    - intx : [xmin, xmax]
    - inty : [ymin, ymax]
    - intt : [tmin, tmax]
    - n : divisions par axe
    - dt : pas de temps du slider

    - routes : liste de routes, chaque route = np.array([[x0,y0],[x1,y1],...])
    - route_times : liste de arrays temps (un par route)
        - si None : on suppose t_k = intt[0] + k*dt
        - sinon : route_times[i] donne le temps de chaque point route[i]
    """

    # -------------------------
    # 1) Grille vent
    # -------------------------
    x_vals = np.linspace(intx[0], intx[1], n + 1)
    y_vals = np.linspace(inty[0], inty[1], n + 1)
    X, Y = np.meshgrid(x_vals, y_vals)
    points = np.column_stack([X.ravel(), Y.ravel()])

    # Temps du slider
    ts = np.arange(intt[0], intt[1] + dt, dt)

    # -------------------------
    # 2) Cache vent
    # -------------------------
    U_cache, V_cache, C_cache = [], [], []

    for t in ts:
        results = np.array([V(p, t) for p in points])
        angles_deg = results[:, 0]
        vitesses = results[:, 1]

        angles_rad = np.radians(angles_deg)
        U = (-vitesses * np.sin(angles_rad)).reshape(X.shape)
        Vv = (-vitesses * np.cos(angles_rad)).reshape(Y.shape)

        C = 1 - np.clip(vitesses / 30, 0, 1)

        U_cache.append(U)
        V_cache.append(Vv)
        C_cache.append(C.flatten())


    # -------------------------
    # 3) Figure
    # -------------------------
    fig, ax = plt.subplots()
    plt.subplots_adjust(bottom=0.25)

    Q = ax.quiver(X, Y, U_cache[0], V_cache[0], color=plt.cm.gray(C_cache[0]))
    ax.set_aspect('equal')
    ax.set_title(f"Temps = {ts[0]:.2f}")

    # -------------------------
    # 4) Routes (optionnel)
    # -------------------------
    route_lines = []
    route_scatters = []
    route_times_list = []

    if routes is not None:
        if route_times is None:
            route_times = [None] * len(routes)

        for k, route in enumerate(routes):
            route = np.asarray(route, dtype=float)

            # temps associés
            if route_times[k] is None:
                rt = intt[0] + np.arange(len(route)) * dt
            else:
                rt = np.asarray(route_times[k], dtype=float)

            route_times_list.append(rt)

            # ligne route (toujours affichée) avec label
            line, = ax.plot(route[:, 0], route[:, 1], lw=2, label=f"Route {k}")
            route_lines.append((line, route))

            # scatter mobile (un point)
            i0 = np.searchsorted(rt, ts[0])
            i0 = np.clip(i0, 0, len(route) - 1)

            sc = ax.scatter(route[i0, 0], route[i0, 1], s=60)
            route_scatters.append((sc, route, rt))

        # ajouter légende après avoir tracé toutes les routes
        ax.legend()


    # -------------------------
    # 5) Slider
    # -------------------------
    ax_slider = plt.axes([0.25, 0.1, 0.5, 0.03])
    slider = Slider(ax_slider, "Temps", intt[0], intt[1], valinit=ts[0], valstep=dt)

    # -------------------------
    # 6) Update
    # -------------------------
    def update(val):
        # index vent
        idx = int(round((slider.val - intt[0]) / dt))
        idx = np.clip(idx, 0, len(ts) - 1)

        # update vent
        Q.set_UVC(U_cache[idx], V_cache[idx])
        Q.set_color(plt.cm.gray(C_cache[idx]))
        ax.set_title(f"Temps = {ts[idx]:.2f}")

        # update routes
        if routes is not None:
            t = ts[idx]
            for sc, route, rt in route_scatters:
                j = np.searchsorted(rt, t)
                j = np.clip(j, 0, len(route) - 1)
                sc.set_offsets([route[j, 0], route[j, 1]])

        fig.canvas.draw_idle()

    slider.on_changed(update)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import time


def vent(V, intx, inty, intt, n, dt=1.0, routes=None, route_times=None):
    """
    Champ de vent avec slider temps + affichage routes.

    Arguments :
    - V : fonction ([x, y], t) -> (angle_deg, vitesse)
    - intx : [xmin, xmax]
    - inty : [ymin, ymax]
    - intt : [tmin, tmax]
    - n : divisions par axe
    - dt : pas de temps du slider

    - routes : liste de routes, chaque route = np.array([[x0,y0],[x1,y1],...])
    - route_times : liste de arrays temps (un par route)
        - si None : on suppose t_k = intt[0] + k*dt
        - sinon : route_times[i] donne le temps de chaque point route[i]
    """

    # -------------------------
    # 1) Grille vent
    # -------------------------
    x_vals = np.linspace(intx[0], intx[1], n + 1)
    y_vals = np.linspace(inty[0], inty[1], n + 1)
    X, Y = np.meshgrid(x_vals, y_vals)
    points = np.column_stack([X.ravel(), Y.ravel()])

    # Temps du slider
    ts = np.arange(intt[0], intt[1] + dt, dt)

    # -------------------------
    # 2) Cache vent
    # -------------------------
    U_cache, V_cache, C_cache = [], [], []

    for t in ts:
        results = np.array([V(p, t) for p in points])
        angles_deg = results[:, 0]
        vitesses = results[:, 1]

        angles_rad = np.radians(angles_deg)
        U = (-vitesses * np.sin(angles_rad)).reshape(X.shape)
        Vv = (-vitesses * np.cos(angles_rad)).reshape(Y.shape)

        C = 1 - np.clip(vitesses / 30, 0, 1)

        U_cache.append(U)
        V_cache.append(Vv)
        C_cache.append(C.flatten())

    # -------------------------
    # 3) Figure
    # -------------------------
    fig, ax = plt.subplots()
    plt.subplots_adjust(bottom=0.25)

    Q = ax.quiver(X, Y, U_cache[0], V_cache[0], color=plt.cm.gray(C_cache[0]))
    ax.set_aspect('equal')
    ax.set_title(f"Temps = {ts[0]:.2f}")

    # -------------------------
    # 4) Routes (optionnel)
    # -------------------------
    route_lines = []
    route_scatters = []
    route_times_list = []

    if routes is not None:
        if route_times is None:
            route_times = [None] * len(routes)

        for k, route in enumerate(routes):
            route = np.asarray(route, dtype=float)

            # temps associés
            if route_times[k] is None:
                rt = intt[0] + np.arange(len(route)) * dt
            else:
                rt = np.asarray(route_times[k], dtype=float)

            route_times_list.append(rt)

            # ligne route (toujours affichée) avec label
            line, = ax.plot(route[:, 0], route[:, 1], lw=2, label=f"Route {k}")
            route_lines.append((line, route))

            # scatter mobile (un point)
            i0 = np.searchsorted(rt, ts[0])
            i0 = np.clip(i0, 0, len(route) - 1)

            sc = ax.scatter(route[i0, 0], route[i0, 1], s=60)
            route_scatters.append((sc, route, rt))

        # ajouter légende après avoir tracé toutes les routes
        ax.legend()
    # -------------------------
    # 5) Slider
    # -------------------------
    ax_slider = plt.axes([0.25, 0.1, 0.5, 0.03])
    slider = Slider(ax_slider, "Temps", intt[0], intt[1], valinit=ts[0], valstep=dt)

    # -------------------------
    # 6) Update (OPTION 2 : throttle)
    # -------------------------
    last_update = 0
    max_fps = 1  # mets 15 si tu veux encore plus léger

    def update(val):
        nonlocal last_update
        now = time.time()

        # throttle (limite la fréquence)
        if now - last_update < 1 / max_fps:
            return
        last_update = now

        # index vent
        idx = int(round((slider.val - intt[0]) / dt))
        idx = np.clip(idx, 0, len(ts) - 1)

        # update vent
        Q.set_UVC(U_cache[idx], V_cache[idx])
        Q.set_color(plt.cm.gray(C_cache[idx]))
        ax.set_title(f"Temps = {ts[idx]:.2f}")

        # update routes
        if routes is not None:
            t = ts[idx]
            for sc, route, rt in route_scatters:
                j = np.searchsorted(rt, t)
                j = np.clip(j, 0, len(route) - 1)
                sc.set_offsets([route[j, 0], route[j, 1]])

        fig.canvas.draw_idle()

    slider.on_changed(update)
    plt.show()
```
